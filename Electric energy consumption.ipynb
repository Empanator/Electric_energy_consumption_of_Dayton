```python
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import kurtosis, skew

%matplotlib inline
```

<div class="alert alert-block alert-info">
<b>Load data and see info</b>
</div>

```python
df = pd.read_csv('DAYTON_hourly.csv')
df.Datetime = pd.to_datetime(df.Datetime)
df.set_index('Datetime', inplace = True)
df.sort_index(inplace = True)
df.info()
```

<div class="alert alert-block alert-info">
<b>Plot original time series and measure skewness</b>
</div>

```python
plt.figure(figsize = (20,10))
plt.rc('font', weight = 'normal')
plt.title('Original Time Series', weight = 'bold', fontsize = 25, pad = 15)
plt.xlabel('Year', weight = 'bold', fontsize = 20, labelpad = 15)
plt.ylabel('Comsumption', weight = 'bold', fontsize = 20,  labelpad = 15)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.tick_params(axis = 'both', pad = 15, size = 10)                                
plt.tight_layout()                                               
plt.rc('font', size = 15)
plt.style.use('ggplot')
plt.plot(df.index, df.DAYTON_MW, color = 'green')
plt.show()

print('skewness is equal to: %f' %skew(df.DAYTON_MW))
```

<div class="alert alert-block alert-info">
<b>Pick a year and month to plot</b>
</div>

```python
year = int(input("Type the year you want from 2004 to 2018: "))
if year > 2003 and year < 2019:
    month = int(input("Type the month you want: "))
    if month > 0 and month < 13:
        plt.figure(figsize = (20,10))
        plt.rc('font', weight = 'normal')
        plt.title('Eletric energy consumption', weight = 'bold', fontsize = 25, pad= 15)
        plt.xlabel('Year-Month-Day', weight = 'bold', fontsize = 20, labelpad = 15)
        plt.ylabel('Consumption', weight = 'bold', fontsize = 20, labelpad = 15)
        plt.xticks(fontsize = 20)
        plt.yticks(fontsize = 20)
        plt.tick_params(axis = 'both', pad = 15, size = 10,)                                
        plt.tight_layout(pad = 5)                                                
        dfyear = df[df.index.year == year]
        filtro = (dfyear.index.month == month)
        plt.style.use('ggplot')
        plt.plot(dfyear[filtro], color = 'green')
else:
    print('Run again and type a valid date')
```

<div class="alert alert-block alert-info">
<b>Pick a year, month and day to plot</b>
</div>

```python
year = int(input("Type the year you want from 2004 to 2018: "))
if year > 2003 and year < 2019:
    month = int(input("Type the month you want: "))
    if month > 0 and month < 13:
        day = int(input("Type the day from the month you want: "))
        if day > 0 and day < 31:
            plt.figure(figsize = (20,10))
            plt.rc('font', weight = 'normal')
            plt.title('Eletric energy consumption', weight = 'bold', fontsize = 25, pad= 15)
            plt.xlabel('Year-Month-Day', weight = 'bold', fontsize = 20, labelpad = 15)
            plt.ylabel('Consumption', weight = 'bold', fontsize = 20, labelpad = 15)
            plt.xticks(fontsize = 20)
            plt.yticks(fontsize = 20)
            plt.tick_params(axis = 'both', pad = 15, size = 10,)                                
            plt.tight_layout(pad = 5)                                                
            dfyear = df[df.index.year == year]
            filtro = (dfyear.index.month == month) & (dfyear.index.day == day) 
            plt.style.use('ggplot')
            plt.plot(dfyear[filtro], color = 'green')
else:
    print('Run again and type a valid date')
```
